```python
G_fb = nx.read_edgelist('F:\\FATIMA FAST\\Courses\\Marketing Analytics\\Practice Exercises\\facebook_combined.txt', create_using = nx.Graph(), nodetype=int)
```

```python
print(nx.info(G_fb))
```

```python
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb);
```

#### We can also visualize the network such that the node color varies with Degree and node size with Betweenness Centrality. The code to do this is:

```python
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```

#### You can also know the labels of the nodes with the highest betweenness centrality using:

```python
sorted(betCent, key=betCent.get, reverse=True)[:5]
```

```python
degree_centrality = nx.degree_centrality(G_fb)
avg_degree_centrality = np.mean(list(degree_centrality.values()))
print("Degree Centrality:", degree_centrality)
print("Average Degree Centrality:", avg_degree_centrality)


```

```python
betweenness_centrality = nx.betweenness_centrality(G_fb)
print("Betweenness Centrality:", betweenness_centrality)
```

```python
avg_betweenness_centrality = np.mean(list(betweenness_centrality.values()))
print("Average Betweenness Centrality:", avg_betweenness_centrality)
```

```python
closeness_centrality = nx.closeness_centrality(G_fb)
print("Closeness Centrality:", closeness_centrality)
```

```python
clustering_coefficient = nx.clustering(G_fb)
print("Clustering Coefficient:", clustering_coefficient)
```

```python

avg_clustering_coefficient = np.mean(list(clustering_coefficient.values()))
print("Average Clustering Coefficient:", format(avg_clustering_coefficient,'.2f'))
```

```python
avg_shortest_path_length = nx.average_shortest_path_length(G_fb)
print("Average Shortest Path Length:", avg_shortest_path_length)
```

```python
diameter = nx.diameter(G_fb)
print("Diameter:", diameter)
```
